# 4 вопрос

Сколько у каждого из пользователей в среднем покупок в неделю (по месяцам)? 

```python
olist['order_month']=olist.order_purchase_timestamp.dt.to_period('M') #колонка с месяцем заказа
```

```python
# считаем количество покупок у каждого пользователя по месяцам и кладём в датафрейм purchases_by_users
purchases_by_users = olist.query('order_status in ["delivered", "shipped", "processing", "invoiced", "approved"]')\
                          .groupby(['customer_unique_id','order_month'], as_index=False).order_id.nunique()\
                          .rename(columns={'order_id': 'purchases_in_month'})
```

```python
# колонка с средним количеством покупок в неделю
purchases_by_users['purchases_by_week']=purchases_by_users.purchases_in_month/purchases_by_users.weeks_count
```

# 5 вопрос

1. Выполните когортный анализ пользователей.

2. В период с января по декабрь выявите когорту с самым высоким retention на 3-й месяц. 

```python
# создаем датафрейм с датой первого заказа для каждого пользователя
first_order = olist.query('order_status in ["delivered", "shipped", "processing", "invoiced", "approved"]')\
                   .groupby('customer_unique_id', as_index=False).order_month.min()\
                   .rename(columns={'order_month':'first_order_date'})

# к каждому заказу присоединяем дату первого заказа пользователя
retention = olist.query('order_status in ["delivered", "shipped", "processing", "invoiced", "approved"]')
retention = retention[['customer_unique_id', 'order_month']].merge(first_order, on='customer_unique_id', how='left')
# для каждого заказа определяем, на какой месяц после первого заказа он был совершён
retention['cohort_period']=(retention.order_month.dt.month-retention.first_order_date.dt.month)+(retention.order_month.dt.year-retention.first_order_date.dt.year)*12
```

```python
# создаем сводную таблицу, где строки - когорта пользователей, столбцы - месяц, значения - retention.
retention = retention.groupby(['first_order_date', 'cohort_period'], as_index=False).customer_unique_id.nunique()\
.rename(columns={'first_order_date':'cohort', 'customer_unique_id':'customers_count'})\
.pivot(index='cohort', columns='cohort_period', values='customers_count')
retention=retention.div(retention[0], axis=0)
```

```python
retention.style.set_caption('').background_gradient(cmap='magma').highlight_null('white').format("{:.2%}", na_rep="")
```

В период с января по декабрь когорта с самым высоким retention на 3-й месяц:
2017: когорта 2017-05(retention=0.40%)
2018: когорта 2018-02(retention=0.30%)

# 6 вопрос

Используя python, построй RFM-сегментацию пользователей, чтобы качественно оценить свою аудиторию. В кластеризации можешь выбрать следующие метрики: R - время от последней покупки пользователя до текущей даты, F - суммарное количество покупок у пользователя за всё время, M - сумма покупок за всё время. Подробно опиши, как ты создавал кластеры. Для каждого RFM-сегмента построй границы метрик recency, frequency и monetary для интерпретации этих кластеров.

```python
current_date = max(olist.order_purchase_timestamp)
F=lambda x: (current_date - max(x)).days
```

```python
RFM =olist.query('order_status in ["delivered", "shipped", "processing", "invoiced", "approved"]')\
          .groupby('customer_unique_id', as_index=False)\
          .agg({'order_purchase_timestamp': F, 'order_id':'nunique', 'price':'sum'})
```

```python
RFM.rename(columns={'order_id': 'Frequency', 'price': 'Monetary', 'order_purchase_timestamp':'Recency'}, inplace=True)
```

```python
plt.figure(figsize=(16,4))
plt.subplot(1,3,1)
sns.histplot(RFM.Recency)
plt.subplot(1,3,2)
sns.histplot(RFM.Frequency)
plt.subplot(1,3,3)
sns.histplot(np.log(RFM.Monetary + 1))  # Добавляем 1, чтобы избежать логарифма нуля
```

Recency и Monetary можно проранжировать с помощью квантилей.
Большая часть пользователей совершила лишь один заказ, поэтому ранжировать Frequency от 1 до 5 бессмыслено, тогда пусть оценка 1 - это пользователи, совершившие один заказ, 2 - пользователи, сделавшие 2 и более заказов.

```python
qR20 = RFM.Recency.quantile(0.2)
qR40 = RFM.Recency.quantile(0.4)
qR60 = RFM.Recency.quantile(0.6)
qR80 = RFM.Recency.quantile(0.8)

def R(x):
    if x < qR20:
        return 5
    elif qR20 <= x < qR40:
        return 4
    elif qR40 <= x < qR60:
        return 3
    elif qR60 <= x < qR80:
        return 2
    else:
        return 1
def F(x):
    if x==1:
        return 1
    else:
        return 2
qM20 = RFM.Monetary.quantile(0.2)
qM40 = RFM.Monetary.quantile(0.4)
qM60 = RFM.Monetary.quantile(0.6)
qM80 = RFM.Monetary.quantile(0.8)
def M(x):
    if x < qM20:
        return 1
    elif qM20 <= x < qM40:
        return 2
    elif qM40 <= x < qM60:
        return 3
    elif qM60 <= x < qM80:
        return 4
    else:
        return 5
```

```python
RFM['R']=RFM.Recency.apply(R)
RFM['F']=RFM.Frequency.apply(F)
RFM['M']=RFM.Monetary.apply(M)
```

```python
RFM['RFM Score'] = RFM['R'].map(str) + RFM['F'].map(str) + RFM['M'].map(str)
```

Поскольку пользователей совершивших лишь одну покупку подавляющее большинство(97%), можно сегментировать только по R и M:
1. "Чемпионы": недавние покупатели с высокими тратами(R=4-5, M=4-5 или Recency=0-226, Monetary=317-13440),
2. "Недавние клиенты": недавние покупатели с низкими тратами(R=4-5, M=1-3 или Recency=0-226, Monetary=0-316),
3. "Требуют внимания": давние покупатели с высокими тратами(R=1-3, M=4-5 или Recency=227-772, Monetary=317-13440),
4. "Заснувшие клиенты": давние покупатели с низкими тратами(R=1-3, M=1-3 или Recency=227-772, Monetary=0-316).

```python
def segment_user(x):
    if x['R'] >=4 and x['M'] >= 4:
        return 'Чемпионы'
    elif x['R'] >= 4 and x['M'] < 4:
        return 'Недавние клиенты'
    elif x['R'] < 4 and x['M'] >= 4:
        return 'Требуют внимания'
    elif x['R'] < 4 and x['M'] < 4:
        return 'Заснувшие клиенты'
```

```python
RFM['Segment'] = RFM.apply(segment_user, axis=1)
```

```python
sns.histplot(RFM.Segment)
plt.xticks(rotation=45) 
```
